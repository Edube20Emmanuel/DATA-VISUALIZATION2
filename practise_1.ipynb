```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("supermarket_sales - Sheet1.csv")
df
```

```python
def exploratory_data_analysis(data):
    print("*"*52,"\n")
    print("\t********EXPLORATORY DATA ANALYSIS******")
    print("\tThe data has %d rows and %d columns"%(df.shape[0],df.shape[1]))
    print("\n\t***DATA TYPES OF OUR DATA SET***")
    for key,value in data.dtypes.items():
        print(f"\t{key}--------------->{value}")
    print("\n\t*****THE COLUMNS IN MY DATA SET*****")
    for column in data.columns:
        print(f"\t{column}")
exploratory_data_analysis(df)
```

```python
## Carring out some beautiful visualization
# Count plot
plt.figure(figsize=(20,10))
sns.countplot(data=df,x="Product line",hue="Gender",palette="dark6",)
```

```python
df.columns
```

```python
## Plotting a bar plot
plt.figure(figsize=(20,10))
sns.barplot(data=df,x="Product line",y='Total',hue="Gender",hue_order=["Male","Female"],estimator=sum,palette="Set2",capsize=0.2)
plt.xticks(rotation = 45)
plt.grid(axis="y")
```

```python
## Plottin box plots
sns.boxplot(data=df,x="City",y="Total",width=0.5,hue="Gender",linewidth=1,linecolor="green",showmeans=True,meanprops={"marker":"*",
                                                                                                                      "markerfacecolor":"white",
                                                                                                                      "markersize":12,
                                                                                                                      "markeredgecolor":"black"})
plt.xticks(rotation=36)
```

```python
df.columns
```

```python
plt.figure(figsize=(20,8))
sns.violinplot(data=df,x='Branch',y='Total',hue='Gender',split=True,cut=True)
```

```python
df.columns
```

```python
plt.figure(figsize=(16,6))
sns.swarmplot(data=df,x='Product line',y='Total',hue='Gender',marker='*')
plt.grid()
```
