```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("supermarket_sales - Sheet1.csv")
df
```

```python
plt.figure(figsize=(12,6))
plt.title("My Barplot Practise")
sns.barplot(x = "Product line",y = "Total",hue = "Gender",hue_order = ["Male","Female"],data = df,palette = "dark",capsize = 0.5,saturation = 5)
```

```python
### Changing the estimator
plt.figure(figsize=(12,6))
plt.title("Sale by Product line in sum")
sns.barplot(x = "Product line",y = "Total",hue = "Gender",data=df,palette = "dark",errorbar=None,
            estimator = sum)
```

```python
plt.figure(figsize=(12,6))
plt.title("Sale by Product line")
sns.barplot(x = "Product line", y = "Total",hue = "Gender",capsize = 0.5,data = df,estimator = np.median,
            palette = "dark", saturation = 5)
```

```python
plt.figure(figsize=(12,6))
plt.title("Sale by Product line")
sns.barplot(x = "Product line", y = "Total",data=df,facecolor=(0,0,0,0),
              linewidth = 3,
              edgecolor = sns.color_palette("dark"),errorbar = None,estimator = sum)
```
