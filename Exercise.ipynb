```python
### Changing columns ' Unit_Cost ',' Unit_Price ', ' Profit ', ' Cost ', 'Revenue' to floats
df[['Unit_Cost','Unit_Price','Profit','Cost','Revenue']] = df[['Unit_Cost','Unit_Price','Profit','Cost','Revenue']].replace("[,$]", "",regex=True).astype('float64')
   
```

```python
df
```

```python
def convert_column_dtypes(data):
    for column in data.columns:
        if column in data.select_dtypes(include="object"):
            data[column] = data[column].astype("category")
        else:
            data[column] = data[column]
    return data.dtypes
```

```python
convert_column_dtypes(df)
```

2. Profitability Analysis (Profit vs Order Quantity)

```python
plt.figure(figsize=(9,5))
sns.lineplot(df,x='Order_Quantity',y='Profit')
plt.title('Profitability Analysis'.upper())
plt.xlabel("Profit")
plt.ylabel("Order_Quantity")
plt.grid()
```

```python
df_silver = df[df['Product_Description'].str.contains('Silver',na=False)]
df_silver
```

```python
df_Black = df[df['Product_Description'].str.contains('Black',na=False)]
df_Black
```

```python
df_Black['Product_Description'] = df_Black['Product_Description'].str.replace('.','Black')
```

```python
df_Black.loc[0:,'Product_Description'] = 'Black'
```

```python
df_Black.reset_index(drop=True)
```

```python
df_silver.loc[0:,'Product_Description'] = 'Silver'
```

```python
df_silver.reset_index(drop=True)
```

```python
New_df = pd.concat([df_Black,df_silver])
```

```python
New_df.reset_index(drop=True)
```

```python
New_df.groupby(["Product_Description"],as_index=None,observed=False)["Revenue"].agg({"sum"})
```

```python
sns.countplot(New_df,x="Product_Description")
```

```python
Revenue_df = New_df.groupby(["Product_Description"],as_index=None,observed=False)["Revenue"].agg({"sum"})
Revenue_df
```

```python
New_df.columns
```

```python
plt.figure(figsize=(8,6))
sns.barplot(New_df,x="Product_Description",y="Revenue",hue="Customer_Gender",capsize=0.2,palette="Set2")
plt.title("Revenue from silver and black Bikes by Gender".title())
plt.xlabel("Bike description by Gender")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.grid(axis='y')
```

```python
plt.figure(figsize=(8,6))
sns.barplot(New_df,x="Product_Description",y="Revenue",hue="Product_Description",capsize=0.5,estimator="sum")
plt.title("Revenue distribution of Black and white bikes".upper())
plt.xticks(rotation = 45)
plt.grid(axis="y")
```
